```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import zipfile

zip_file_path = 'diwali.zip'
csv_file_name_in_zip = 'Python_Diwali_Sales_Analysis/Diwali Sales Data.csv'

with zipfile.ZipFile(zip_file_path, 'r') as z:
    with z.open(csv_file_name_in_zip) as f:
        df = pd.read_csv(f, encoding='latin1')
```

```python
df.drop(['Status' , 'unnamed1'],axis = 1,inplace = True)
```

**Exploratiry Data Analysis**

```python
df.columns
```

**Gender Group of people**

```python
genderx = sns.countplot(x = 'Gender' , data = df)
plt.xlabel('gender')
plt.ylabel('count')
plt.title('count of gender')
for bars in genderx.containers:
    genderx.bar_label(bars)
plt.show()
```

```python
df.head(3)
```

**gender by amount they spend **

```python
amount_they_spend = df.groupby('Gender')['Amount'].sum().sort_values(ascending = False)
ax = sns.barplot(x = amount_they_spend.index,y = amount_they_spend.values)
plt.xlabel('gender')
plt.ylabel('amount')
plt.title('gender by amount they spend')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
ax = sns.countplot(x = 'Age Group' , data = df , hue='Gender')
plt.xlabel('age group')
plt.ylabel('count')
plt.title('count of age group')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

**#total amount vs age group**

```python
df.head()
```

```python
sales_age_group = df.groupby('Age Group')['Amount'].sum().sort_values(ascending = False)
ax = sns.barplot(x = sales_age_group.index,y = sales_age_group.values)
plt.xlabel('age group')
for bars in ax.containers:
    ax.bar_label(bars)
```

**total number of orders from top 10 states**

```python
df.head()
```

```python
sales_state = df.groupby(['State'])['Orders'].sum().sort_values(ascending = False).head(10)

plt.figure(figsize=(25,10))
ax = sns.barplot(x = sales_state.index,y = sales_state.values)
plt.xlabel('state')
plt.ylabel('orders')
plt.title('total number of orders')
plt.xticks(rotation=45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

**total amount/sales from top 10 states**

```python
df.head(2)
```

```python
total_sales_state = df.groupby('State')['Amount'].sum().sort_values(ascending = False).head(10)

plt.figure(figsize=(25,10))
plt.xticks(rotation=45)
ax = sns.barplot(x = total_sales_state.index,y = total_sales_state.values)
plt.xlabel('state')
plt.ylabel('amount')
```

**Marital status**

```python
ax = sns.countplot(x = 'pelli aipoindhi' , data = df)
plt.xlabel('pelli aipoindhi')
plt.ylabel('count')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesby_marital = df.groupby(['pelli aipoindhi', 'Gender'])['Amount'].sum().sort_values(ascending = False)
salesby_marital_df = salesby_marital.reset_index()

plt.figure(figsize=(10,5))
ax = sns.barplot(x = 'pelli aipoindhi', y = 'Amount', hue = 'Gender', data = salesby_marital_df)
plt.xlabel('pelli aipoindhi')
plt.ylabel('amount')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
