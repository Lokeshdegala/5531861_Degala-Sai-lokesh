```python
df['order_date'] = pd.to_datetime(df['order_date'],dayfirst=True)
df['day_name'] = df['order_date'].dt.day_name()
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_name'] = pd.Categorical(df['day_name'], categories=weekday_order, ordered=True)

order_by_day = df.groupby('day_name', observed=False)['order_id'].nunique()
ax = order_by_day.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.title('Daily Trend of Pizza Orders')
plt.show()
```

**% of sales by category**

```python
df['quantity'].sum()
```

```python
category_sales = df.groupby('pizza_category')['total_price'].sum()
category_sales_percentage = category_sales / category_sales.sum() * 100
plt.figure(figsize=(10,6))
plt.pie(category_sales_percentage, labels=category_sales_percentage.index, autopct = '%1.1f%%' , startangle=90, wedgeprops={'edgecolor': 'black', 'width' : 0.2})
plt.title('Percentage of Sales by Category')
```

**% sales by pizza size & category **

```python
df.head(2)
```

```python
sales_category_by_pizzasize = df.groupby(['pizza_size','pizza_category'])['total_price'].sum()
sales_percentage_total = sales_category_by_pizzasize / sales_category_by_pizzasize.sum() * 100
pivot_table = sales_percentage_total
pivot_table.plot(kind='bar',figsize=(10,6))
plt.xlabel('Pizza Size')
plt.ylabel('Percentage of Sales')
plt.title('Percentage of Sales by Pizza Size and Category')
```

```python
sales_category_by_pizzasize = df.groupby(['pizza_size','pizza_category'])['total_price'].sum()
sales_percentage_total = sales_category_by_pizzasize / sales_category_by_pizzasize.sum() * 100
pivot_table = sales_percentage_total.unstack()
pivot_table.plot(kind='bar',stacked=True, figsize=(10,6))
plt.xlabel('Pizza Size')
plt.ylabel('Percentage of Sales')
plt.title('Percentage of Sales by Pizza Size and Category')
```

**Top 5 best selling pizzas **

```python
df.head(2)
```

```python
best_top5_pizzas = df.groupby('pizza_name')['total_price'].sum()
best_top5_pizzas = best_top5_pizzas.sort_values(ascending=False).head(5)
plt.figure(figsize=(10,9))
plt.bar(best_top5_pizzas.index, best_top5_pizzas.values)
plt.xlabel('Pizza Name')
plt.ylabel('Total Pizzas Sold')
plt.xticks(rotation=45)
plt.title('Top 5 Best Selling Pizzas')
```
