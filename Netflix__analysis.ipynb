```python
import numpy as np
import pandas as pd
import plotly.express as px
from textblob import TextBlob

df = pd.read_csv('netflix_titles.csv.zip')
```

```python
x = df.groupby(['rating']).size().reset_index(name='counts')
print(x)
```

```python
piechart = px.pie(x , values = 'counts' , names = 'rating' , title = 'distribution of content ratings on netflix')
piechart.show()
```

```python
df['director'] = df['director'].fillna('No Director Specified')
```

```python
df.head()
```

```python
directors_list = pd.DataFrame()
print(directors_list)
```

```python
directors_list = df['director'].str.split(',',expand = True).stack()
print(directors_list)
```

```python
directors_list = directors_list.to_frame()
print(directors_list)
```

```python
directors_list.columns = ['Director']
print(directors_list)
```

```python
directors = directors_list.groupby(['Director']).size().reset_index(name='counts')
print(directors)
```

```python
directors = directors[directors.Director != 'No Director Specified']
```

```python
print(directors)
```

```python
directors = directors.sort_values(by = ['counts'], ascending = False)

print(directors)
```

```python
top5directors = directors.head()
print(top5directors)
```

```python
top5directors = top5directors.sort_values(by = ['counts'],ascending = False)

barchart = px.bar(top5directors , x = 'counts' , y = 'Director' , title = 'Top 5 Directors on Netflix')
barchart.show()
```

```python
df['cast'] = df['cast'].fillna('no cast specified')
cast_df = pd.DataFrame()
cast_df = df['cast'].str.split(',', expand = True).stack()
cast_df = cast_df.to_frame()
cast_df.columns = ['Actor']
actors = cast_df.groupby(['Actor']).size().reset_index(name = 'counts')
actors = actors[actors.Actor != 'no cast specified']
actors = actors.sort_values(by = ['counts'], ascending = False)
top5actors = actors.head()
top5actors = top5actors.sort_values(by = ['counts'])
barchart2 = px.bar(top5actors , x = 'counts' , y = 'Actor' , title = 'Top 5 Actors on Netflix')
barchart2.show()
```

```python
df1 = df[['type', 'release_year']]
df1 = df1.rename(columns = {'release_year': 'release_year', "type": 'Type'})
df2 = df1.groupby(['release_year', 'Type']).size().reset_index(name = "counts")
```

```python
print(df2)
```

```python
df2 = df2[df2['release_year']>=2000]
graph = px.line(df2, x = 'release_year', y = 'counts', color = 'Type' , title = 'trend of content produced on netflix')
graph.show()
```

```python
df['country'] = df['country'].fillna('not specified')
country_df = pd.DataFrame()
country_df = df['country'].str.split(',' , expand = True).stack()
country_df = country_df.to_frame()
country_df.columns = ['Country']
Country = country_df.groupby(['Country']).size().reset_index(name = 'counts')
Country = Country[Country['Country'] != 'not specified']
Country = Country.sort_values('counts' , ascending = False)
top5Country = Country.head()
top5Country = top5Country.sort_values('counts')
graph2 = px.bar(top5Country , x = 'counts' , y = 'Country' ,title = 'Top 5 Countries on Netflix')
graph2.show()
```

```python
df3 = df[['release_year','description']]
df3 = df3.rename(columns = {'release_year': 'release year' , 'description': 'Description'})
for index,row in df3.iterrows():
    d=row['Description']
    testimonial = TextBlob(d)
    p = testimonial.sentiment.polarity
    if p==0:
      sent = 'Neutral'
    elif p>0:
      sent = 'Positive'
    else:
      sent = 'Negative'
    df3.loc[[index,2],'sentiment']=sent
df3 = df3.groupby(['release year','sentiment']).size().reset_index(name = 'counts')
df3 = df3[df3['release year']>=2005]
bargraph = px.bar(df3 , x = 'release year' , y = 'counts' , color = 'sentiment' , title = 'sentiment of content on netflix')
bargraph.show()
```
