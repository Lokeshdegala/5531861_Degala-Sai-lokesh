Q4 ) Does price varies with airlines ?


```python
df.head()
```

```python
df.groupby('airline')['price'].mean()
```

```python
import seaborn as sns
plt.pyplot.figure(figsize = (10,5))
plt.pyplot.xticks(rotation = 90)

# Group by both airline and class, then calculate the mean price for each combination
grouped_data = df.groupby(['airline', 'class'])['price'].mean().reset_index()

# Now use the grouped data for plotting, with 'class' as the hue
ax = sns.barplot(x = 'airline' , y = 'price' , hue = 'class', data=grouped_data)
for bars in ax.containers:
    ax.bar_label(bars)
plt.pyplot.show()
```

```python
plt.pyplot.figure(figsize = (10,5))
plt.pyplot.xticks(rotation = 90)
ax =  sns.barplot(x = df['airline'] , y = df['price'] , hue = df['class'])

plt.pyplot.show()
```

```python
plt.pyplot.figure(figsize = (10,5))
plt.pyplot.xticks(rotation = 90)
grouped_data =  df.groupby(['airline' , 'class'])['price'].mean().reset_index()
ax =  sns.barplot(x = 'airline' , y = 'price' , hue = 'class', data=grouped_data)

plt.pyplot.show()
```
