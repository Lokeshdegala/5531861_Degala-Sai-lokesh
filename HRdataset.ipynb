**Q15) IDENTIFY DEPARTMENTS WITH THE HIGHEST ATTRITION RATE (RESIGNED %)**

```python
df.head(2)
```

```python
attrition_rate = df.groupby('Department')['Status'].agg(total_emp = 'count' , resigned_emp = lambda x : (x == 'Resigned').sum())
```

```python
attrition_rate['attrition_rate_%'] = (attrition_rate['resigned_emp'] / attrition_rate['total_emp']) * 100
```

```python
attrition_rate.sort_values('attrition_rate_%' , ascending = False)
```
