```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat', 'low fat' : 'Low Fat', 'reg' : 'Regular'})
```

```python
df['Item Fat Content'].unique()
```

**KPI'S Requirement **

```python
total_sales = df['Sales'].sum()
```

```python
avg_sales = df['Sales'].mean()
```

```python
no_of_items_sold = df['Sales'].count()
```

```python
avg_ratings = df['Rating'].mean()
```

```python
print(f"total_sales : ${total_sales:,.1f}")
print(f"avg_sales : ${avg_sales}")
print(f"no_of_items_sold : {no_of_items_sold}")
print(f"avg_ratings : {avg_ratings}")
```

**chart's requirement**

```python
#total sales by fat content

sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct='%.1f%%', startangle=90)
plt.axis('equal')
plt.title('Total Sales by Fat Content')
plt.show()
```

```python
df.head(2)
```

```python
#total sales by item type

sales_by_item_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 8))
plt.xticks(rotation=90)
bars = plt.bar(sales_by_item_type.index, sales_by_item_type.values)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')
for bar in bars:
  plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() ,
           f'{bar.get_height():,.0f}', ha='center', va='bottom')
plt.show()
```

```python
df.head(2)
```

```python
#fat content by oulet for total sales

grouped = df.groupby(['Item Fat Content','Outlet Location Type'])['Sales'].sum().unstack()
grouped = grouped[['Tier 1', 'Tier 2', 'Tier 3']]

ax = grouped.plot(kind='bar', figsize=(12,5), title='Fat Content by Outlet for Total Sales')
plt.xlabel('Item Fat Content')
plt.ylabel('Total Sales')
plt.legend(title='Outlet Location Type')
plt.xticks(rotation=90)
plt.show()
```

```python
df.head(2)
```

```python
#total sales by establisment
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(8,4))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o' ,linestyle='-')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year')
for x,y in zip(sales_by_year.index, sales_by_year.values):
  plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom')
plt.show()
```

```python
df.head(2)
```

```python
#sales by outlet size

groupsed = df.groupby('Outlet Size')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(9,4))
plt.pie(groupsed, labels=groupsed.index, autopct= '%1.1f%%', startangle = 90)
plt.tight_layout()
plt.title('Sales by Outlet Size')
plt.show()
```

```python
df.head(2)
```

```python
#sales by outlet location
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales',ascending=False)

plt.figure(figsize=(8,3))
ax = sns.barplot(x='Sales' , y='Outlet Location Type', data=sales_by_location)

for index, value in enumerate(sales_by_location['Sales']):
    plt.text(value, index, f'{value:,.0f}', va='center')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')
plt.title('Sales by Outlet Location Type')
plt.show()
```
